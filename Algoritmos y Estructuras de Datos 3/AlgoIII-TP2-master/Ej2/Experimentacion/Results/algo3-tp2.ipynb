### Correlacion

```python
df_Dijkstra = Dijkstra[['n','m','tiempo']].groupby(['n','m']).mean().reset_index()
df_Dijkstra_FIFO = Dijkstra_FIFO[['n','m','tiempo']].groupby(['n','m']).mean().reset_index()
df_Bellman_Ford = Bellman_Ford[['n','m','tiempo']].groupby(['n','m']).mean().reset_index()
df_Floyd_Warshall = Floyd_Warshall[['n','m','tiempo']].groupby(['n','m']).mean().reset_index()
#print(df_Dijkstra)
#print(df_Dijkstra_FIFO)
#print(df_Bellman_Ford)
#print(df_Floyd_Warshall)
#Cotas
df_Dijkstra['cota'] = df_Dijkstra['n']*df_Dijkstra['m']*np.log(df_Dijkstra['n'])
df_Dijkstra_FIFO['cota'] = df_Dijkstra_FIFO['n']*df_Dijkstra_FIFO['m']*np.log(df_Dijkstra_FIFO['n'])
df_Bellman_Ford['cota'] = (df_Bellman_Ford['n'] ** 2) * df_Bellman_Ford['m']
df_Floyd_Warshall['cota'] = df_Floyd_Warshall['n'] ** 3

#print(df_Floyd_Warshall)

```

```python
ax1 = sns.lmplot(x='tiempo', y='cota', data=df_Dijkstra);

r = np.corrcoef(df_Dijkstra["tiempo"], df_Dijkstra["cota"])[0, 1];
plt.title('Coeficiente:'+str(r))
plt.xlabel("T(n,m)");
plt.ylabel("cota(n,m)");

print("El coeficiente de correlación de Pearson es: r="+str(r))
```
